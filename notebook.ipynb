## 5. Upload your PDF

Or skip this cell to use `input/AlisaDocs.pdf` from the repo.

```python
from google.colab import files
import shutil, os
os.makedirs('input', exist_ok=True)
uploaded = files.upload()
INPUT_FILE = list(uploaded.keys())[0] if uploaded else 'AlisaDocs.pdf'
if uploaded:
    shutil.move(INPUT_FILE, f'input/{INPUT_FILE}')
print('Using:', f'input/{INPUT_FILE}')
```

## 6. Run the cleaner

```python
OUTPUT_FILE = INPUT_FILE.rsplit('.', 1)[0] + '_clean.pdf'
!python -m src.pipeline input/{INPUT_FILE} output/{OUTPUT_FILE} --debug
```

## 7. Download the result

```python
from google.colab import files
files.download(f'output/{OUTPUT_FILE}')
```
